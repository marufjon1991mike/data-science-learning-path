```python
import pandas as pd
df = pd.read_csv("final_data_2023.csv")
df = df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
df.describe()
```

```python
yaqin_sharq = df[df['country'].isin(["United Arab Emirates", "Saudi Arabia", "Qatar", "Kuwait", "Bahrain", "Oman"])]
yaqin_sharq
```

```python
# task_1 gdp ustuni bo'yicha eng yuqori 5 ta mamlakatni toping (eng boy mamlakatlar).
df.sort_values("gdp", ascending = False).head(5)
```

```python
# task_2 gdp ustuni bo'yicha eng past 5 ta mamlakatni toping (eng kambag'al mamlakatlar, ya'ni GDP eng kichik).
df.sort_values("gdp").head(5)
```

```python
# task_3 rate ustuni bo'yicha saralab, natijadan faqat country va rate ustunlarini ko'rsating (butun jadval emas, faqat shu ikkita ustun):
saralangan = df.sort_values("rate")
saralangan[["country", "rate"]]
```

```python
# task_4 Ikkita shartni birlashtiring — avval continent == "AS" (Osiyo) bo'yicha filtrlang, so'ng natijani rate bo'yicha kamayish tartibida saralang, va faqat birinchi 5 tasini ko'rsating.
osiyo = df[df["continent"] == "AS"]
osiyo_saralangan = osiyo.sort_values("rate", ascending=False).head(5)
osiyo_saralangan
```

```python
# task_5 Har bir qit'a bo'yicha o'rtacha gdp ni hisoblang.
df.groupby("continent")["rate"].mean()
```

```python
# task_6 Har bir qit'ada nechta mamlakat borligini sanang (.count() yoki .size() ishlatib ko'ring, ikkalasini ham sinab, farqini payqang).
df.groupby("continent")["country"].count()
```

```python
df.groupby("continent")["country"].size()
```

```python
# task_7 .agg() yordamida, har bir qit'a uchun rate ustunining mean, min, max larini bitta jadvalda ko'rsating:
df.groupby("continent")["rate"].agg(["mean","min", "max"])
```

```python
# task_8 faqat oecd == 1 bo'lgan mamlakatlarni filtrlang, so'ng ularni continent bo'yicha guruhlab, o'rtacha rate ni toping.
saralangan_oecd = df[df['oecd'] == 1]
saralangan_oecd.groupby("continent")["rate"].mean()
```
