```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
df['Age'] = df['Age'].fillna(df["Age"].median()) # Age qatoridagi bo'sh qiymatlarga median qiymatlarni qo'shadi
```

```python
df['Cabin'] = df['Cabin'].fillna("Noma'lum")
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0]) # mode — statistikada "eng ko'p takrorlanuvchi qiymat" degani
```

```python
df.drop(columns=["Ticket"]) # "Ticket" ustunini o'chirib tashlash
```

```python
df.drop(index=[0,1]) # 0 va 1 indexdagi qatorlarni o'chirib tashlash
```

```python
df["OilaSoni"] = df["SibSp"] + df["Parch"] # Bu — SibSp (aka-uka/er-xotin soni) va Parch (ota-ona/farzand soni) ustunlarini qo'shib, yangi "oila a'zolari soni" ustunini yaratadi.
```

```python
# task_16 .map() yordamida, Sex ustunidagi "male"/"female" qiymatlarini "Erkak"/"Ayol" ga o'zgartirib, yangi JinsUZ ustuniga saqlang.
df["JinsUz"] = df["Jins"].map(lambda x : "Erkak" if x=="male" else "Ayol")
```

```python
df
```

```python
# task_17 .apply() yordamida, Age asosida yangi YoshToifasi ustunini yarating: 18 dan kichik bo'lsa "Bola", 18-60 orasida "Kattalar", 60 dan katta bo'lsa "Keksa".
df["yoshToifasi"] = df["Age"].apply(lambda x: "Bola" if x<18 else ("Kattalar" if x<60 else "Keksalar"))
```

```python
df
```

```python
# task_18 Alohida funksiya yozing, Fare asosida chipta narxini toifalarga ajratadi:
def narx_toifasi(narx):
  if narx<10:
    return "Arzon"
  elif narx<50:
    return "O'rta"
  else:
    return "Qimmat"
df["NarxToifasi"] = df["Narx"].apply(narx_toifasi)
```

```python
df
```

```python
pd.crosstab(df["Toifa"], df["NarxToifasi"]) # Bu — ikkita kategoriyaviy ustun orasidagi bog'liqlikni ko'rsatadigan jadval (masalan, "1-toifadagi yo'lovchilarning qanchasi haqiqatan ham 'Qimmat' toifasiga tushadi").
```

```python
# task_20 Name ustunidan familiyani ajratib oling (vergulgacha bo'lgan qism):
df["Familiya"] = df["Name"].str.split(",").str[0]
```

```python
# task_21 Name ustunida "Mrs." so'zi bor-yo'qligini tekshirib, yangi TurmushQurgan (Boolean) ustunini yarating:
df["TurmushQurgan"] = df["Name"].str.contains("Mrs.")
```

```python
mijozlar = pd.DataFrame({
    "mijoz_id": [1, 2, 3],
    "ism": ["Ali", "Vali", "Sami"]
})

buyurtmalar = pd.DataFrame({
    "buyurtma_id": [101, 102, 103],
    "mijoz_id": [1, 2, 1],
    "summa": [50000, 30000, 20000]
})
```

```python
# task_24 Yuqoridagi mijozlar va buyurtmalar DataFrame'larini yarating, pd.merge() bilan birlashtiring, natijani ko'ring.
natija = pd.merge(mijozlar, buyurtmalar, on="mijoz_id")
```

```python
natija
```

```python
# task_25 mijozlar ga yana bir mijoz qo'shing — {"mijoz_id": 4, "ism": "Nodira"} — lekin unga hech qanday buyurtma bermang. Endi how="left" bilan birlashtiring — Nodira natijada ko'rinadimi? summa ustunida uning qatorida nima chiqadi?
mijozlar = pd.DataFrame({
    "mijoz_id": [1, 2, 3, 4],
    "ism": ["Ali", "Vali", "Sami", "Nodira"]
})

pd.merge(mijozlar, buyurtmalar, on="mijoz_id", how="left")
```

```python
# task_26 Xuddi shu ikkita jadvalni how="inner" bilan birlashtiring — Nodira bu safar ko'rinadimi? Farqni tushuntiring.
pd.merge(mijozlar, buyurtmalar, on="mijoz_id", how="inner")
```

```python
# task_29 Sun'iy sana ustuni yarating va undan yil/oy/kun ajratib oling:
sanalar = pd.DataFrame({"sana": ["2023-01-15", "2023-06-20", "2024-03-10"]})
sanalar
```

```python
sanalar["kun"] = sanalar["sana"].dt.day
sanalar
```
