```python
# ==============================================
# NETFLIX KONTENT TAHLILI
# Maqsad: Netflixdagi film va seriallar bo'yicha
# umumiy tendensiyalarni aniqlash
# ==============================================
import pandas as pd

# Ma'lumotni yuklaymiz
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
# --------Ma'lumot sifatini tekshirish--------
# bo'sh qiymatlar qancha va qaysi ustunlarda ekanini ko'ramiz
print(df.isna().sum())
print("\nBo'sh qiymatlar foizda")
print(df.isna().sum() / len(df) *100)
```

```python
# director ustunida bo'sh qiymatlar asosan TV Show turidagi kontentga tegishli
# (chunki serialda bir nechta epizod rejissyori bo'lishi mumkin, bitta emas)
# Shuning uchun bu tabiiy holat, "Noma'lum" bilan belgilaymiz
df["director"] = df['director'].fillna("Noma'lum")
df.head()
```

```python
df[df['cast'].isna()]["type"].value_counts()
```

```python
# cast bo'shligi biror bitta turga aniq bog'liq emas (Movie va TV Show deyarli teng taqsimlangan),
# shuning uchun bu tasodifiy bo'shliq deb hisoblaymiz va "Noma'lum" bilan belgilaymiz
df["cast"] = df['cast'].fillna("Noma'lum")
df.head()
```

```python
# country ustunini eng ko'p takrorlangan mamlakat nomi bilan to'ldiramiz
df["country"] = df["country"].fillna(df["country"].mode()[0])
```

```python
# date_added bu taxmin qilish qiyin bo'lgan ma'lumaot, xatolik ehtimoli ko'p bo'lganligi uchun bo'sh qatorlarni o'chirib tashlaymiz
df = df.dropna(subset="date_added")
df.head()
```

```python
# rating ustunidagi bo'sh qiymatlarni eng ko'p takrorlangan qiymat bilan to'ldiramiz
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
df[df["duration"].isna()]
```

```python
# duration ustunida mavjud bo'lmagan qiymatlar rating ustunidagi mos qatorlarga o'tib qolgan. Qiymatlarni to'g'ri joyga ko'chiramiz
mask = df["duration"].isna()
df.loc[mask, "duration"] = df.loc[mask, "rating"]
```

```python
# rating ustunidan ko'chirilgan ma'lumotlarni (74 min, 84 min, 66 min) eng ko'p takrorlangan rating bilan almashtiramiz
df.loc[mask, "rating"] = df["rating"].mode()[0]
df.loc[[5541, 5794, 5813]]
```

```python
# show_id — shunchaki identifikator, tahliliy ma'no bermaydi, olib tashlaymiz
df = df.drop(columns="show_id")
df.head()
```

```python
# Har bir kontent nechta janrga tegishli ekanini hisoblaymiz —
# bu keyinchalik "ko'p janrli kontent ko'proq mashhurmi" kabi savollarga yordam beradi
df["JanrSoni"] = df["listed_in"].str.split(",").str.len()
df.head()
```

```python
# Kontentni chiqarilgan yiliga qarab "Eski" va "Yangi" toifalarga bo'lamiz —
# bu Netflix'ning so'nggi yillardagi tendensiyasini ko'rish uchun foydali
df["Eski_yoki_yangi"] = df["release_year"].apply(lambda x: "Yangi" if x>=2015 else "Eski")
df.head()
```

```python
df.shape
```

```python
# Yillar bo'yicha, Netflix qaysi turdagi kontentga (Movie/TV Show) ko'proq e'tibor bergan? (Maslahat: date_added dan yilni ajratib, groupby yoki
# pivot_table bilan type bo'yicha taqsimotni ko'ring)
# pivot_table yordamida har bir yilda qaysi kontentga ko'proq e'tibor berilganini ko'ramiz
```

```python
# date_added ichidan yillarni ajratib olamiz
# date_added ichida str formatni sanaga o'zgartiramiz
df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"])

# Sanadan faqat yilni ajratib olamiz
df["Qo'shilgan_yil"] = df["date_added"].dt.year
```

```python
pivot = df.pivot_table(values="title", index="Qo'shilgan_yil", columns="type", aggfunc="count")
pivot["Nisbat"] = pivot["Movie"] / pivot["TV Show"]
pivot
```

```python
# -------- Xulosa --------
# Netflix har doim Movie'ga TV Show'dan ko'proq e'tibor bergan (nisbat doimiy ravishda 2:1 dan 2.4:1 orasida), lekin 2016-2019 yillar oralig'ida ikkalasi
# ham keskin o'sgan — bu Netflix'ning eng faol kontent yaratish davri bo'lgan. 2020-2021 yillarda esa pasayish ko'rinadi (1424→1284→993 Movie uchun),
# bu ehtimol COVID-19 pandemiyasi ta'sirida ishlab chiqarish sekinlashgani bilan bog'liq bo'lishi mumkin — bu qiziqarli, tekshirish mumkin bo'lgan gipoteza.
```

```python
# Qaysi 5 ta davlat eng ko'p kontent ishlab chiqargan? (Maslahat: value_counts() yodingizdami?)
df["country"].value_counts().head()
```

```python
# Eng ko'p uchraydigan 5 ta janr qaysilar? (Diqqat: listed_in da bir nechta janr bitta katakchada, vergul bilan — buni avval qanday ajratganingizni eslang,
# keyin har bir janrni alohida sanash kerak bo'ladi — bu biroz qiyinroq, agar qiynalsangiz ayting)
df["listed_in"] = df["listed_in"].str.split(",")
```

```python
# rating (masalan, TV-MA, PG-13) bilan Eski_yoki_yangi orasida qanday bog'liqlik bor — masalan, yangi kontentlarda qattiqroq reytinglar ko'proqmi?
# (Maslahat: pivot_table yoki crosstab)
pd.crosstab(df['Eski_yoki_yangi'], df['rating'])
```
