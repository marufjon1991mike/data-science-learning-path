```python
df["director"] = df["director"].fillna("Noma'lum") # director — MNAR (Missing Not At Random) (TV Show bilan bog'liq), "Noma'lum" bilan to'ldiramiz
```

```python
df["cast"] = df["cast"].fillna("Noma'lum") #  cast — kam foiz, matn ustuni, "Noma'lum" mantiqiy
```

```python
df["country"] = df["country"].fillna(df["country"].mode()[0]) # country — kam foiz, eng ko'p uchraydigan davlat bilan to'ldiramiz
```

```python
df["rating"] = df["rating"].fillna(df["rating"].mode()[0]) # rating — juda kam bo'sh, mode bilan to'ldirish mantiqiy
```

```python
df = df.dropna(subset=("date_added")) # date_added — sana, va foiz kam, shuning uchun bu qatorlarni olib tashlaymiz
```

```python
df[df["duration"].isna()]
```

```python
mask = df["duration"].isna() # bu 3 ta muammoli qatorni belgilaydi (True/False ustuni)
df.loc[mask, "duration"] = df.loc[mask, "rating"] # o'sha 3 qatorda, ratingdagi qiymatni ("74 min") durationga ko'chiramiz
df.loc[mask, "rating"] =  df["rating"].mode()[0] # endi rating bo'shab qoldi, uni eng ko'p uchraydigan reyting bilan to'ldiramiz
```

```python
df.isna().sum()
```

```python
# task_1 show_id ustunini olib tashlang (bu — shunchaki identifikator, tahlil uchun foydasiz — nega ekanini o'zingizga izohlang).
df = df.drop(columns=["show_id"])
df.head(5)
```

```python
# task_2 listed_in ustuni (janrlar, vergul bilan ajratilgan, masalan "Comedies, Dramas") asosida, yangi JanrSoni ustunini yarating — bu har bir film/serial nechta janrga tegishli ekanini ko'rsatsin (maslahat: .str.split(",") va uzunligini olish kerak bo'ladi — buni qanday birlashtirish mumkinligini o'ylab ko'ring).
df["JanrSoni"] = df["listed_in"].str.split(",").str.len()
df.head(5)
```

```python
# task_3 release_year asosida yangi Eski_yoki_yangi ustunini yarating — agar release_year >= 2015 bo'lsa "Yangi", aks holda "Eski" (shartli mantiq — 3-haftadagi bilimingizni eslang).
df["Eski_yoki_yangi"] = df['release_year'].map(lambda x: "Yangi" if x >= 2015 else "Eski")
```

```python
df['Eski_yoki_yangi'].value_counts()
```

```python
# task_4 Yangi qator qo'shing — o'zingiz o'ylab topgan, hayoliy bitta film/serial ma'lumoti bilan (barcha ustunlarni to'ldirib).
yangi_qator = pd.DataFrame([{"type" : "Movie", "title" : "Qaroqchilar", "director" : "Ahad Qayum", "cast" : "Noma'lum", "country":"Uzbekistan", "date_added":"July 23, 2026", "release_year":2026, "rating":"Noma'lum", "duration":"70 min", "listed_in":"Fantastic", "description":"Noma'lum"}])
df = pd.concat([df, yangi_qator], ignore_index=True)
```

```python
# task_5 type, title, country ustunlarini mos ravishda Turi, Sarlavha, Davlat deb qayta nomlang (bitta rename() chaqiruvida).
df = df.rename(columns={"type":"Turi", "title":"Sarlavha", "country":"Davlat"})
```

```python
# task_6 title (yoki yangi nomlangan Sarlavha) ustunini indeks qilib belgilang, natijani ko'ring.
df = df.set_index("Sarlavha")
df.head(5)
```

```python
# task_7 Indeksni bekor qilib, yana oddiy raqamli indeksga qaytaring.
df = df.reset_index()
df.head(5)
```
