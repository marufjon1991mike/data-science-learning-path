```python
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
df.head()
```

```python
# bo'sh qiymatlarni tekshiramiz
df.isna().sum()
```

```python
# yangi hisob-toifasi ustuni yaratamiz
df["HisobToifasi"] = df["total_bill"].apply(lambda x: "Arzon" if x<20 else ("O'rta" if x<40 else "Qimmat"))
df
```

```python
# choychaqa foizini hisoblovchi yangi ustun yaratamiz
df["Foiz"] = df["tip"] / df["total_bill"] * 100
df
```

```python
# kun bo'yicha o'rtacha choychaqa foizini topamiz
# natijani kamayish tartibida joylashtiramiz
df.pivot_table(values="Foiz", index="day", aggfunc="mean").sort_values("Foiz", ascending=False)
```

```python
# Erkaklar va ayollar qoldiradigan o'rtacha choychaqa farini tekshiramiz
df.pivot_table(values="tip", index="sex", aggfunc="mean")
```

```python
# chekuvchi/chekmaydigan, kun bo'yicha buyurtmalar sonini tekshiramiz
df.pivot_table(values="total_bill", index="day", columns="smoker", aggfunc="count")
```

```python
# eng yuqori choychaqa foizini (Foiz ustuni) bergan 5 ta mijozning barcha ma'lumotlarini topamiz
# Foiz ustunini yuqoridan pastga saralaymiz. eng yuqorida eng katta qiymat qoladi
df.sort_values("Foiz", ascending=False).head()
```
